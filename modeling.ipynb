```python
#importing required packages from sklearn 
!pip3 install catboost
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegressionCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import VotingClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score 
from catboost import CatBoostClassifier
```

```python
#importing datasets
all_data=pd.read_csv('/content/all_data.csv')  # all_data 
labels=pd.read_csv('/content/0bf8bc6e-30d0-4c50-956a-603fc693d966.csv') # labels for train data
train=pd.read_csv('/content/4910797b-ee55-40a7-8668-10efd5c1b960.csv')  # training data
test=pd.read_csv('/content/702ddfc5-68cd-4d1d-a0de-f5f566f76d91.csv') # test data
```

```python
all_data = all_data.drop(['id', 'subvillage', 'lga', 'ward','recorded_by'], axis=1) #dropping irrelevent columns
categorical = ['funder',
               'installer',
               'basin',
               'scheme_management',
               'extraction_type_class',
               'management',
               'management_group',
               'payment_type',
               'water_quality',
               'quantity',
               'source',
               'source_class',
               'waterpoint_type','permit','quantity_group']
for i in categorical:
    all_data[i] = pd.factorize(all_data[i])[0] #fatorzing categorical data in zeros and one
```

```python
labels = pd.factorize(labels.status_group)[0] #factorzing label column 
```

```python
train_X= all_data[0:len(train)]
test_X = all_data[len(train):]
```

```python
# Dividing data into training and testing data
X_train,X_test,y_train,y_test=train_test_split(train_X,labels,test_size=.3,random_state=69) 
```

```python
# training logisticregressionCV model 
model_lr=LogisticRegressionCV(cv=5)
model_lr.fit(X_train,y_train)
prediction_lr=model_lr.predict(X_test)
accuracy_score(prediction_lr,y_test)
```

```python
# training baseline CatBoostClassifier model on training data
model_cbc = CatBoostClassifier()
model_cbc.fit(X_train,y_train)
prediction_cbc=model_cbc.predict(X_test)
accuracy_score(y_test,prediction_cbc)
#0.7948372615039282
```

```python
# training baseline RandomForestClassifier model on training data
model_rf=RandomForestClassifier()
model_rf.fit(X_train,y_train)
prediction_rf=model_rf.predict(X_test)
accuracy_score(prediction_rf,y_test)
```

```python
# using cross_val_score to cross validate randomforest model 
score = cross_val_score(model_rf, train_X, labels, scoring='accuracy', cv=3).mean()
print('3 fold cross-validation score:', score.round(4))
```

```python
#training basline xgboost model 
xgb = XGBClassifier(max_depth=10, n_estimators=300, silent=False, booster='gbtree')

# Fit the model
xgb.fit(X_train, y_train)

# make predictions for test data
y_pred = xgb.predict(X_test)

# evaluate predictions
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')

# Print the results
print("Accuracy: %.2f%%" % (accuracy * 100.0))
print("Precision: %.2f%%" % (precision * 100.0))
```

```python
score = cross_val_score(xgb, train_X, labels, scoring='accuracy', cv=3).mean()
print('3 fold cross-validation score:', score.round(4))
```

```python
#training baseline extratree model
model_etc=ExtraTreesClassifier()
model_etc.fit(X_train,y_train)
y_pred=model_etc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')

# Print the results
print("Accuracy: %.2f%%" % (accuracy * 100.0))
print("Precision: %.2f%%" % (precision * 100.0))
```

```python
# stacking three best performing models
from sklearn.ensemble import StackingClassifier
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
def get_stacking():
	# define the base models
	level0 = list()
	level0.append(('rf',RandomForestClassifier(n_estimators=500,max_depth=20)))
	level0.append(('cat',CatBoostClassifier(iterations=1000, loss_function='MultiClass',
                         bootstrap_type='Bernoulli', boosting_type='Plain', 
                         max_depth=8, random_state=42, od_type="Iter")))
	level0.append(('xgb', XGBClassifier(max_depth=10, n_estimators=300, silent=True, booster='gbtree',min_child_weight=10,gamma=.5,reg_alpha=.1)))
	# define meta learner model
	level1 = SVC()
	# define the stacking ensemble
	model = StackingClassifier(estimators=level0, final_estimator=level1, cv=3)
	return model

# get a list of models to evaluate
def get_models():
	models = dict()
	models['stacking'] = get_stacking()
	return models

# evaluate a give model using cross-validation
def evaluate_model(model, X, y):
	scores = cross_val_score(model, X, y, scoring='accuracy', cv=3, n_jobs=-1, error_score='raise')
	return scores

# get the models to evaluate
models = get_models()
# evaluate the models and store results
results, names = list(), list()
for name, model in models.items():
	scores = evaluate_model(model, X_train, y_train)
	results.append(scores)
	names.append(name)
	print(name, scores.mean())
```
