```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Exploration

```python
labels=pd.read_csv('/content/0bf8bc6e-30d0-4c50-956a-603fc693d966.csv')
all_data=pd.read_csv('/content/all_data.csv')
```

```python
plot_df=all_data.loc[all_data.origin=='train']
```

```python
plot_df = plot_df.merge(labels, how='inner', on='id')
```

```python
plot = plot_df.groupby(['region_code', 'status_group']) \
               .quantity \
               .count() \
               .unstack()

plot.plot(kind='bar', stacked=True, figsize=(12, 10), rot=40);
```
